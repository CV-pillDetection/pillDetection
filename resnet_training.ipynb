```python
import os
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from PIL import Image, ImageOps
from tqdm import tqdm
import numpy as np
import cv2
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# ì„¤ì •
# ==========================================
IMAGE_DIR = '/content/drive/MyDrive/ResNet Dataset'
CSV_PATH = '/content/drive/MyDrive/solopill/drug_info_unique_drug_id_final.csv'
SAVE_PATH = '/content/drive/MyDrive/pill_shape_color_resnet50_color_fixed.pth'
BATCH_SIZE = 32
LEARNING_RATE = 0.0005
NUM_EPOCHS = 40
IMG_SIZE = 128
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Shape ë§¤í•‘
SHAPE_MAPPING = {
    'ì›í˜•': 0,
    'ì¥ë°©í˜•': 1,
    'íƒ€ì›í˜•': 2,
    'ê¸°íƒ€': 3, 'íŒ”ê°í˜•': 3, 'ë§ˆë¦„ëª¨í˜•': 3, 'ë°˜ì›í˜•': 3,
    'ì‚¬ê°í˜•': 3, 'ì •ì‚¬ê°í˜•': 3, 'ì‚¼ê°í˜•': 3, 'ì˜¤ê°í˜•': 3, 'ìœ¡ê°í˜•': 3
}

# Color ë§¤í•‘
COLOR_MAPPING = {
    'í•˜ì–‘': 0, 'íˆ¬ëª…': 0,
    'ë…¸ë‘': 1, 'ì—°ë‘': 1,
    'ë¶„í™': 2, 'ìì£¼': 2, 'ë¹¨ê°•': 2,
    'ì£¼í™©': 3,
    'íŒŒë‘': 4, 'ë‚¨ìƒ‰': 4, 'ì²­ë¡': 4,
    'ê°ˆìƒ‰': 5, 'íšŒìƒ‰': 5, 'ê²€ì •': 5,
    'ì´ˆë¡': 6, 'ë³´ë¼': 6
}

NUM_SHAPES = 4
NUM_COLORS = 7

# í´ë˜ìŠ¤ ì´ë¦„
SHAPE_NAMES = ['ì›í˜•', 'ì¥ë°©í˜•', 'íƒ€ì›í˜•', 'ê¸°íƒ€']
COLOR_NAMES = ['í°ìƒ‰', 'ë…¸ë‘', 'í•‘í¬', 'ì˜¤ë Œì§€', 'ë¸”ë£¨', 'ë¸Œë¼ìš´', 'ê·¸ë¦°']

# ==========================================
# Label Smoothing
# ==========================================
class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self, smoothing=0.1, weight=None, ignore_index=-100):
        super().__init__()
        self.smoothing = smoothing
        self.weight = weight
        self.ignore_index = ignore_index

    def forward(self, pred, target):
        n_class = pred.size(1)
        mask = target != self.ignore_index
        target = target.clone()
        target[~mask] = 0

        one_hot = torch.zeros_like(pred).scatter(1, target.unsqueeze(1), 1)
        smooth_one_hot = one_hot * (1 - self.smoothing) + self.smoothing / n_class

        log_probs = torch.log_softmax(pred, dim=1)
        loss = -(smooth_one_hot * log_probs).sum(dim=1)

        if self.weight is not None:
            weight_mask = self.weight[target]
            loss = loss * weight_mask

        loss = loss * mask.float()

        return loss.sum() / mask.float().sum().clamp(min=1.0)

# ==========================================
# ë°ì´í„°ì…‹ í´ë˜ìŠ¤ (Train/Val ë¶„ë¦¬)
# ==========================================
class PillCSVDataset(Dataset):
    def __init__(self, img_dir, csv_path, transform=None, split='train', val_ratio=0.15):
        self.img_dir = img_dir
        self.transform = transform

        print(f"ë°ì´í„° ë§¤ì¹­ ì‹œì‘ ({split} set)...")
        df = pd.read_csv(csv_path)

        self.shape_dict = dict(zip(df['drug_id'], df['drug_shape']))
        self.color_dict = dict(zip(df['drug_id'], df['color_class1']))

        self.valid_file_list = []
        self.shape_labels = []
        self.color1_labels = []

        all_files = os.listdir(img_dir)

        for filename in tqdm(all_files):
            if not filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                continue
            try:
                file_id = filename.split('_')[0]
            except:
                continue

            if file_id in self.shape_dict and file_id in self.color_dict:
                shape_text = self.shape_dict[file_id]
                color_text = self.color_dict[file_id]

                if shape_text not in SHAPE_MAPPING:
                    continue
                shape_label = SHAPE_MAPPING[shape_text]

                colors = [c.strip() for c in color_text.split(',')]
                if len(colors) == 0 or colors[0] not in COLOR_MAPPING:
                    continue

                color1 = COLOR_MAPPING[colors[0]]

                self.valid_file_list.append(filename)
                self.shape_labels.append(shape_label)
                self.color1_labels.append(color1)

        # Train/Val ë¶„ë¦¬
        np.random.seed(42)
        indices = np.random.permutation(len(self.valid_file_list))
        val_size = int(len(indices) * val_ratio)

        if split == 'train':
            indices = indices[val_size:]
        else:  # validation
            indices = indices[:val_size]

        self.valid_file_list = [self.valid_file_list[i] for i in indices]
        self.shape_labels = [self.shape_labels[i] for i in indices]
        self.color1_labels = [self.color1_labels[i] for i in indices]

        self.shape_counts = [0] * NUM_SHAPES
        self.color_counts = [0] * NUM_COLORS
        for s, c in zip(self.shape_labels, self.color1_labels):
            self.shape_counts[s] += 1
            self.color_counts[c] += 1

        print(f"âœ… {split} ë§¤ì¹­ ì™„ë£Œ: {len(self.valid_file_list)}ì¥")
        print(f"   Shape ë¶„í¬: {self.shape_counts}")
        print(f"   Color ë¶„í¬: {self.color_counts}")

    def __len__(self):
        return len(self.valid_file_list)

    def __getitem__(self, idx):
        filename = self.valid_file_list[idx]
        shape_label = self.shape_labels[idx]
        color_label = self.color1_labels[idx]

        img_path = os.path.join(self.img_dir, filename)
        try:
            image = Image.open(img_path).convert('RGB')
        except:
            image = Image.new('RGB', (128, 128), (255, 255, 255))

        if self.transform:
            image = self.transform(image)

        return image, shape_label, color_label

    def get_class_weights(self, counts):
        weights = []
        total = sum(counts)

        for count in counts:
            if count > 0:
                weight = np.log(total / count + 1) + 1.0
                weights.append(weight)
            else:
                weights.append(1.0)

        weights = torch.FloatTensor(weights)
        weights = weights / weights.mean()
        weights = torch.clamp(weights, min=0.5, max=1.5)

        return weights

# ==========================================
# ëª¨ë¸ ì •ì˜
# ==========================================
class ShapeColorResNet(nn.Module):
    def __init__(self, num_shapes, num_colors):
        super(ShapeColorResNet, self).__init__()

        resnet = models.resnet50(weights='IMAGENET1K_V1')

        for param in resnet.parameters():
            param.requires_grad = False

        self.features = nn.Sequential(*list(resnet.children())[:-1])
        num_ftrs = resnet.fc.in_features

        self.fc_shape = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(num_ftrs, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_shapes)
        )

        self.fc_color = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(num_ftrs, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_colors)
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        shape_out = self.fc_shape(x)
        color_out = self.fc_color(x)
        return shape_out, color_out

    def unfreeze_last_layers(self, num_layers=2):
        children = list(self.features.children())
        for child in children[-num_layers:]:
            for param in child.parameters():
                param.requires_grad = True

# ==========================================
# ì „ì²˜ë¦¬
# ==========================================
class PadToSquare:
    def __call__(self, img):
        w, h = img.size
        max_wh = max(w, h)
        padding = ((max_wh - w)//2, (max_wh - h)//2,
                   (max_wh - w+1)//2, (max_wh - h+1)//2)
        return ImageOps.expand(img, padding, fill=255)

train_transform = transforms.Compose([
    PadToSquare(),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.15, hue=0.05),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

val_transform = transforms.Compose([
    PadToSquare(),
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# ==========================================
# í‰ê°€ í•¨ìˆ˜ (Precision, Recall, F1 ê³„ì‚°)
# ==========================================
def evaluate_model(model, dataloader, device):
    model.eval()

    all_shape_preds = []
    all_shape_labels = []
    all_color_preds = []
    all_color_labels = []

    with torch.no_grad():
        for inputs, shape_labels, color_labels in tqdm(dataloader, desc="Evaluating"):
            inputs = inputs.to(device)

            shape_out, color_out = model(inputs)

            _, shape_pred = torch.max(shape_out, 1)
            _, color_pred = torch.max(color_out, 1)

            all_shape_preds.extend(shape_pred.cpu().numpy())
            all_shape_labels.extend(shape_labels.numpy())
            all_color_preds.extend(color_pred.cpu().numpy())
            all_color_labels.extend(color_labels.numpy())

    # Shape ì§€í‘œ ê³„ì‚°
    shape_precision, shape_recall, shape_f1, _ = precision_recall_fscore_support(
        all_shape_labels, all_shape_preds, average='weighted', zero_division=0
    )
    shape_acc = np.mean(np.array(all_shape_preds) == np.array(all_shape_labels))

    # Color ì§€í‘œ ê³„ì‚°
    color_precision, color_recall, color_f1, _ = precision_recall_fscore_support(
        all_color_labels, all_color_preds, average='weighted', zero_division=0
    )
    color_acc = np.mean(np.array(all_color_preds) == np.array(all_color_labels))

    metrics = {
        'shape': {
            'accuracy': shape_acc * 100,
            'precision': shape_precision,
            'recall': shape_recall,
            'f1_score': shape_f1,
            'predictions': all_shape_preds,
            'labels': all_shape_labels
        },
        'color': {
            'accuracy': color_acc * 100,
            'precision': color_precision,
            'recall': color_recall,
            'f1_score': color_f1,
            'predictions': all_color_preds,
            'labels': all_color_labels
        }
    }

    return metrics

# ==========================================
# Confusion Matrix ì‹œê°í™”
# ==========================================
def plot_confusion_matrix(y_true, y_pred, class_names, title, save_path=None):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.title(title, fontsize=16, fontweight='bold')
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"âœ… Confusion matrix ì €ì¥: {save_path}")
    plt.show()

# ==========================================
# ìƒì„¸ ë¦¬í¬íŠ¸ ì¶œë ¥
# ==========================================
def print_detailed_report(metrics, split='Validation'):
    print(f"\n{'='*60}")
    print(f"{split} Set - ìƒì„¸ ì„±ëŠ¥ ì§€í‘œ")
    print(f"{'='*60}")

    print(f"\nğŸ“Š Shape (ëª¨ì–‘) ë¶„ë¥˜:")
    print(f"   Accuracy:  {metrics['shape']['accuracy']:.2f}%")
    print(f"   Precision: {metrics['shape']['precision']:.4f}")
    print(f"   Recall:    {metrics['shape']['recall']:.4f}")
    print(f"   F1-Score:  {metrics['shape']['f1_score']:.4f}")

    print(f"\nğŸ¨ Color (ìƒ‰ìƒ) ë¶„ë¥˜:")
    print(f"   Accuracy:  {metrics['color']['accuracy']:.2f}%")
    print(f"   Precision: {metrics['color']['precision']:.4f}")
    print(f"   Recall:    {metrics['color']['recall']:.4f}")
    print(f"   F1-Score:  {metrics['color']['f1_score']:.4f}")

    print(f"\n{'='*60}")

    print("\nğŸ“‹ Shape í´ë˜ìŠ¤ë³„ ìƒì„¸ ë¦¬í¬íŠ¸:")
    print(classification_report(
        metrics['shape']['labels'],
        metrics['shape']['predictions'],
        target_names=SHAPE_NAMES,
        zero_division=0
    ))

    print("\nğŸ“‹ Color í´ë˜ìŠ¤ë³„ ìƒì„¸ ë¦¬í¬íŠ¸:")
    print(classification_report(
        metrics['color']['labels'],
        metrics['color']['predictions'],
        target_names=COLOR_NAMES,
        zero_division=0
    ))

# ==========================================
# í•™ìŠµ
# ==========================================
def train():
    # ë°ì´í„°ì…‹ ìƒì„±
    train_dataset = PillCSVDataset(IMAGE_DIR, CSV_PATH, transform=train_transform, split='train')
    val_dataset = PillCSVDataset(IMAGE_DIR, CSV_PATH, transform=val_transform, split='val')

    if len(train_dataset) == 0:
        print("âŒ ë°ì´í„° ì—†ìŒ!")
        return

    train_loader = DataLoader(
        train_dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=2,
        pin_memory=True
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )

    model = ShapeColorResNet(NUM_SHAPES, NUM_COLORS).to(DEVICE)

    shape_weights = train_dataset.get_class_weights(train_dataset.shape_counts).to(DEVICE)
    color_weights = train_dataset.get_class_weights(train_dataset.color_counts).to(DEVICE)

    criterion_shape = LabelSmoothingCrossEntropy(smoothing=0.1, weight=shape_weights)
    criterion_color = LabelSmoothingCrossEntropy(smoothing=0.1, weight=color_weights)

    optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.01)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS, eta_min=LEARNING_RATE * 0.01)

    max_grad_norm = 1.0
    best_val_f1 = 0.0
    history = {
        'train_shape_acc': [], 'train_color_acc': [],
        'val_shape_acc': [], 'val_color_acc': [],
        'val_shape_f1': [], 'val_color_f1': []
    }

    print(f"\nğŸ”¥ í•™ìŠµ ì‹œì‘ (Epochs: {NUM_EPOCHS})...")

    for epoch in range(NUM_EPOCHS):
        # ============ Training ============
        model.train()
        running_loss = 0.0
        shape_correct = 0
        color_correct = 0
        total = 0

        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}")

        for inputs, shape_labels, color_labels in pbar:
            inputs = inputs.to(DEVICE)
            shape_labels = shape_labels.to(DEVICE)
            color_labels = color_labels.to(DEVICE)

            optimizer.zero_grad()

            shape_out, color_out = model(inputs)

            loss_shape = criterion_shape(shape_out, shape_labels)
            loss_color = criterion_color(color_out, color_labels)
            loss = loss_shape + 1.2 * loss_color

            if torch.isnan(loss) or torch.isinf(loss):
                continue

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_grad_norm)
            optimizer.step()

            running_loss += loss.item()

            _, shape_pred = torch.max(shape_out, 1)
            _, color_pred = torch.max(color_out, 1)

            total += shape_labels.size(0)
            shape_correct += (shape_pred == shape_labels).sum().item()
            color_correct += (color_pred == color_labels).sum().item()

            pbar.set_postfix({
                'loss': f'{loss.item():.4f}',
                'shape_acc': f'{100*shape_correct/total:.1f}%',
                'color_acc': f'{100*color_correct/total:.1f}%'
            })

        train_shape_acc = 100 * shape_correct / total
        train_color_acc = 100 * color_correct / total

        # ============ Validation ============
        val_metrics = evaluate_model(model, val_loader, DEVICE)

        # íˆìŠ¤í† ë¦¬ ì €ì¥
        history['train_shape_acc'].append(train_shape_acc)
        history['train_color_acc'].append(train_color_acc)
        history['val_shape_acc'].append(val_metrics['shape']['accuracy'])
        history['val_color_acc'].append(val_metrics['color']['accuracy'])
        history['val_shape_f1'].append(val_metrics['shape']['f1_score'])
        history['val_color_f1'].append(val_metrics['color']['f1_score'])

        # ì¶œë ¥
        print(f"\n{'='*60}")
        print(f"Epoch {epoch+1}/{NUM_EPOCHS}")
        print(f"{'='*60}")
        print(f"Train - Shape: {train_shape_acc:.2f}% | Color: {train_color_acc:.2f}%")
        print(f"Val   - Shape: {val_metrics['shape']['accuracy']:.2f}% (F1={val_metrics['shape']['f1_score']:.4f}) | "
              f"Color: {val_metrics['color']['accuracy']:.2f}% (F1={val_metrics['color']['f1_score']:.4f})")

        # Best ëª¨ë¸ ì €ì¥ (F1 í‰ê·  ê¸°ì¤€)
        avg_f1 = (val_metrics['shape']['f1_score'] + val_metrics['color']['f1_score']) / 2
        if avg_f1 > best_val_f1:
            best_val_f1 = avg_f1
            torch.save({
                'epoch': epoch + 1,
                'model_state_dict': model.state_dict(),
                'val_metrics': val_metrics,
                'history': history
            }, SAVE_PATH.replace('.pth', '_best.pth'))
            print(f"âœ… Best ëª¨ë¸ ì €ì¥ (í‰ê·  F1={avg_f1:.4f})")

        scheduler.step()

        # Fine-tuning
        if epoch == 14:
            print("\nğŸ”“ Fine-tuning ì‹œì‘")
            model.unfreeze_last_layers(num_layers=2)
            optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE * 0.1, weight_decay=0.01)
            scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS - 15, eta_min=LEARNING_RATE * 0.001)

    # ============ ìµœì¢… í‰ê°€ ============
    print("\n\n" + "="*60)
    print("ìµœì¢… í‰ê°€ (Best Model)")
    print("="*60)

    # Best ëª¨ë¸ ë¡œë“œ
    checkpoint = torch.load(SAVE_PATH.replace('.pth', '_best.pth'), weights_only=False)
    model.load_state_dict(checkpoint['model_state_dict'])

    final_metrics = evaluate_model(model, val_loader, DEVICE)
    print_detailed_report(final_metrics, split='Validation')

    # Confusion Matrix ì €ì¥
    plot_confusion_matrix(
        final_metrics['shape']['labels'],
        final_metrics['shape']['predictions'],
        SHAPE_NAMES,
        'Shape Classification - Confusion Matrix',
        SAVE_PATH.replace('.pth', '_shape_cm.png')
    )

    plot_confusion_matrix(
        final_metrics['color']['labels'],
        final_metrics['color']['predictions'],
        COLOR_NAMES,
        'Color Classification - Confusion Matrix',
        SAVE_PATH.replace('.pth', '_color_cm.png')
    )

    # ìµœì¢… ì €ì¥
    torch.save({
        'model_state_dict': model.state_dict(),
        'final_metrics': final_metrics,
        'history': history
    }, SAVE_PATH)

    print(f"\nğŸ’¾ ëª¨ë¸ ì €ì¥ ì™„ë£Œ: {SAVE_PATH}")

if __name__ == '__main__':
    train()
```
